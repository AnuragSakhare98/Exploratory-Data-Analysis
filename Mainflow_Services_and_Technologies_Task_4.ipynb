# Visualization

## Bar Plot of 'Total Video Publish Count Per Year'

```python
# Extract the year from the publish_time
df['year'] = df['publish_time'].dt.year

# Group by year and count the number of video_id per year
yearly_counts = df.groupby('year')['video_id'].count()

# Create a bar plot with different colors for each bar
fig, ax = plt.subplots(figsize=(12, 6))
colors = plt.cm.tab20.colors  # Use a colormap with enough colors
bars = ax.bar(yearly_counts.index, yearly_counts.values, color=colors[:len(yearly_counts)])

# Add labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Total Publish Count')
ax.set_title('Total Video Publish Per Year')

# Display the plot
plt.show()
```

## Bar Plot of 'Total Views per Year'

```python
# Group by year and sum the views for each year
yearly_views = df.groupby('year')['views'].sum()

# Create a bar chart with different colors for each bar and set the figure size
plt.figure(figsize=(12, 6))
colors = plt.cm.tab20.colors  # Use a colormap with enough colors
bars = plt.bar(yearly_views.index, yearly_views.values, color=colors[:len(yearly_views)])

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Total Views')
plt.title('Total Views per Year')
plt.xticks(rotation=0)
plt.tight_layout()

# Show the bar chart
plt.show()
```

## Bar Plot of 'Top 5 Categories by Views'

```python
# Group the data by 'category_name' and calculate the sum of 'views' in each category
category_views = df.groupby('category_name')['views'].sum().reset_index()

# Sort the categories by views in descending order and get the top 5
top_categories = category_views.sort_values(by='views', ascending=False).head(5)

# Create a bar plot to visualize the top 5 categories with different colors
plt.figure(figsize=(12, 6))
colors = plt.cm.tab20.colors  # Use a colormap with enough colors
bars = plt.bar(top_categories['category_name'], top_categories['views'], color=colors[:5])

# Add labels and title
plt.xlabel('Category Name', fontsize=12)
plt.ylabel('Total Views', fontsize=12)
plt.title('Top 5 Categories by Views', fontsize=15)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the bar chart
plt.show()
```

## Bar Plot of 'Video Count by Category'

```python
# Create the figure and specify the size
plt.figure(figsize=(12, 6))

# Generate a count plot with different colors for each bar
palette = sns.color_palette("tab20", df['category_name'].nunique())  # Ensure enough colors for each category
sns.countplot(x='category_name', data=df, order=df['category_name'].value_counts().index, hue='category_name', palette=palette, dodge=False, legend=False)

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Add the title
plt.title('Video Count by Category')

# Add grid lines with black color
plt.grid(True, axis='y', linestyle='--', linewidth=0.7, color='black')

# Adjust layout to fit everything
plt.tight_layout()

# Display the plot
plt.show()
```

## Visualizing Box Plot with Outliers for Numeric Columns

```python
# Select numeric columns for outlier detection
numeric_cols = ['views', 'likes', 'dislikes', 'comment_count']

# Plot box plots with outliers
plt.figure(figsize=(13, 6))
df[numeric_cols].boxplot(showfliers=True)
plt.title('Box Plot with Outliers')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.show()
```

## Handling Outliers using the IQR Method and Visualizing Box Plot after Outlier Handling

```python
# Define a function to handle outliers using IQR method
def handle_outliers_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = df[col].apply(lambda x: min(max(x, lower_bound), upper_bound))
    return df

# Handle outliers for each numeric column using IQR method
for col in numeric_cols:
    df = handle_outliers_iqr(df, col)


# Plot box plots after handling outliers
plt.figure(figsize=(22, 6))  # Adjust the width and height as needed
plt.subplot(1, 2, 2)
df[numeric_cols].boxplot()
plt.title('Box Plot after Handling Outliers')

plt.tight_layout()
plt.show()
```
